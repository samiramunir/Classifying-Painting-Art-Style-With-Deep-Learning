```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import numpy as np
def binary_data(target):
    categories = ['cubism', 'expressionism', 'impressionism', 'modern', 'surrealism']
    if target in categories:
        
        categories.remove(target) # remove target category from list
        
        path = 'data/'+target
        target_generator = ImageDataGenerator(rescale=1./255).flow_from_directory( 
        path, target_size=(512, 512), batch_size = 1000, seed = 123) 
        
        target_images, target_labels = next(target_generator)
        other_images = []
        
        for category in categories:
            newpath = 'data/'+category
            # generate 250 images from each other category
            new_generator = ImageDataGenerator(rescale=1./255).flow_from_directory( 
            newpath, target_size=(512, 512), batch_size = 400, seed = 123) 
            other_images.append(next(new_generator)[0])
        
        return target_images, np.concatenate(other_images)
    
    else:
        return "Category Not Found"
```

```python
import talos
from keras.optimizers import Adam, sgd
from keras.activations import relu, tanh, sigmoid
from keras.losses import binary_crossentropy

p = {'lr': [1, 3, 5, 7],
     'dropout': [0.1],
     'optimizer': [Adam],
     'activation':[relu],
    'batch_size': [2]}

```

```python
from keras import models
from keras import layers
from keras import regularizers
from talos.model.normalizers import lr_normalizer

def painting_classifier(x_train, y_train, x_val, y_val, params):

    # next we can build the model exactly like we would normally do it
    model = models.Sequential()
   
    
    model.add(layers.Conv2D(16, (5, 5), strides = 5, input_shape=(256,256,3),
                    activation=params['activation']))
    
    model.add(layers.Conv2D(32, (5, 5), strides = 5,
                    activation=params['activation']))
    
    model.add(layers.Conv2D(32, (3, 3), strides = 5,
                    activation=params['activation']))
    
    model.add(layers.Flatten())
    
    model.add(layers.Dropout(params['dropout']))
    
    model.add(layers.Dense(1024,
                    activation=params['activation'],
                    kernel_initializer='normal'))
    
   
    # then we finish again with completely standard Keras way
    model.add(layers.Dense(1, activation=sigmoid))

    
    model.compile(loss=binary_crossentropy,
                  # here we add a regulizer normalization function from Talos
                  optimizer=params['optimizer'](lr=lr_normalizer(params['lr'],params['optimizer'])),
                  metrics=['acc'])
    
    out = model.fit(x_train, y_train, 
                        validation_data=[x_test, y_test],
                        batch_size=20,
                        epochs=50,
                        verbose=0)

    return out, model
```

```python
results2 = talos.Scan(x_train, y_train, params=p, model=painting_classifier, grid_downsample=0.5)
```
